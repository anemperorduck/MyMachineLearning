```python
# step

class NB_Guassian:
    def __init__(self):
        self.model = None
        self.class_priors = None    # 先验概率
        self.class_means = None     # 每个类别的特征均值
        self.class_vars = None      # 每个类别的特征方差
        self.n_classes = None       # 类别数量
        self.n_features = None      # 特征数量
    
    def fit(self, X, y):
        tmp_set = set(y)
        self.n_classes = len(tmp_set)
        self.n_features = X.shape[1]

        # 初始化参数
        self.class_priors = np.zeros(self.n_classes)
        self.class_means = np.zeros((self.n_classes, self.n_features))
        self.class_vars = np.zeros((self.n_classes, self.n_features))

        # 计算参数
        for i in range(self.n_classes):
            X_i = X[y==i]

            # 先验概率
            self.class_priors[i] = float(len(X_i))/ float(len(X))
            # 均值
            self.class_means[i] = np.mean(X_i, axis=0)
            # 方差
            self.class_vars[i] = np.var(X_i, axis=0, ddof=1)     # ddof 无偏估计
            self.class_vars[i] = np.maximum(self.class_vars[i], 1e-9)     # 不为0
        
        return self
    
    def predict(self, pre_data):     
        # 初始化
        log_probs = np.zeros((1, self.n_classes))

        # 计算
        for i in range(self.n_classes):
            # 似然概率
            log_likehood = -0.5 * np.sum(np.log(2*np.pi*self.class_vars[i]) + ( (pre_data - self.class_means[i])**2 / self.class_vars[i] ))
        
            log_probs[:,i] = np.log(self.class_priors[i]) + log_likehood
        
        return np.argmax(log_probs), log_probs

    def get_params(self):
        return {
            'class_priors': self.class_priors,
            'class_means': self.class_means,
            'class_vars': self.class_vars,
            'n_classes': self.n_classes,
            'n_features': self.n_features
        }

    def print_params(self):
        params = self.get_params()
        for key in params.keys():
            print(f"{key}:\n{params[key]}")    
        print("\n")
```

```python
# step

iris = load_iris()
data, target = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(data, target ,test_size=0.2)

nbg = NB_Guassian()
nbg.fit(X_train, y_train)


sample_index = np.random.randint(0, len(X_test))
test_sample = X_test[sample_index]
true_label = y_test[sample_index]

pre_label, probs = nbg.predict(test_sample)

nbg.print_params()
print(f"预测概率(log)：{probs}")
print(f"预测结果：{pre_label}")
print(f"实际结果：{true_label}")
```
