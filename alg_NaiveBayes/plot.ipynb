```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from collections import Counter
import math
```

```python
iris = load_iris()
data = iris.data
target = iris.target
sepal_length = data[:,0]
sepal_width = data[:,1]
petal_length = data[:, 2]
petal_width = data[:, 3]
```

```python
figure_index_0 = np.where(target==0)[0]
figure_index_1 = np.where(target==1)[0]
figure_index_2 = np.where(target==2)[0]
```

```python
tmp = sepal_length[figure_index_0]
print(figure_index_0.shape)
print(tmp.shape)
tmp
```

```python
plt.figure(figsize=(10, 8))

plt.scatter(sepal_length[figure_index_0], sepal_width[figure_index_0], 
            color='red', label='Setosa (0)', alpha=0.7, s=60)
plt.scatter(sepal_length[figure_index_1], sepal_width[figure_index_1],
            color='green',label='Versicolor (1)', alpha=0.7, s=60)
plt.scatter(sepal_length[figure_index_2], sepal_width[figure_index_2],
            color='blue',label='Versicolor (1)', alpha=0.7, s=60)
```

```python
plt.figure(figsize=(10, 8))

plt.scatter(petal_length[figure_index_0], petal_width[figure_index_0], 
            color='red', label='Setosa (0)', alpha=0.7, s=60)
plt.scatter(petal_length[figure_index_1], petal_width[figure_index_1],
            color='green',label='Versicolor (1)', alpha=0.7, s=60)
plt.scatter(petal_length[figure_index_2], petal_width[figure_index_2],
            color='blue',label='Versicolor (1)', alpha=0.7, s=60)
```
